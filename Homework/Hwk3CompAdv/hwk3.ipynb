```python
import matplotlib.pyplot as plt

# Data for Country W
w_food = [9000, 0]
w_clothes = [0, 3000]

# Data for Country Z
z_food = [37500, 0]
z_clothes = [0, 12500]

plt.figure(figsize=(10, 7))

# Plotting the PPFs
plt.plot(w_clothes, w_food, label='Country W (F = 9000 - 3C)', color='blue', linewidth=2)
plt.plot(z_clothes, z_food, label='Country Z (F = 37500 - 3C)', color='red', linewidth=2)

# Adding labels and title
plt.title('Production Possibilities Frontiers (PPF) - GLBL 5010', fontsize=14)
plt.xlabel('Clothing (units)', fontsize=12)
plt.ylabel('Food (units)', fontsize=12)

# Set axes to start at zero
plt.xlim(0, 14000)
plt.ylim(0, 40000)

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Annotate the parallel nature
plt.text(4000, 30000, 'Parallel Slopes: m = -3', fontsize=10, bbox=dict(facecolor='white', alpha=0.5))

# Save the plot
plt.savefig('PPF_Graph.png', dpi=300)
print("Graph saved as PPF_Graph.png")
plt.show()
```
